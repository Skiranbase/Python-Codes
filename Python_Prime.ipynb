# Prime
A prime number is a whole number greater than 1 whose only factors are 1 and itself.

Problem Description
Write a program to find Nth prime number in a given range, inclusive of the first and the last element in that range. Program should accept input on console. In case of a bad input, program should output appropriate message as mentioned in output specification below.

The problem should obey following constraints:

Valid range for this problem statement is between 0 and 1500000 (Fifteen Lakh / One and a half million)

First element of the range should be less than the last element in the range

Prime number index starts from 1 e.g.

Third prime in range 1 to 10 is 5

Fifth prime in range 1 to 15 is 11

Program should be optimized such that it executes in least amount of time

A prime number (or a prime) is a natural number greater than 1 that has no positive divisors other than 1 and itself.

Output
For Valid inputs

Prime Number at Nth position in a given range, followed by a new line

For Invalid input

Invalid Input, followed by a new line

OR

No prime number is present at this index, followed by a new line

as is applicable.

Input
Line 1 From Number, say N1

Line 2 To Number, say N2 

Line 3 Nth prime number between N1 and N2 where N >=1

Output
For Valid inputs

Prime Number at Nth position in a given range, followed by a new line

For Invalid input

Invalid Input, followed by a new line

OR

No prime number is present at this index, followed by a new line

as is applicable.

Constraints
None

Complexity
Medium

Time Limit
1

Examples
Example 1

Input

1

15

6

Output

13

Example 2

Input

1

15

9

Output

No prime number is present at this index

Example 3

Input

59

96

0

Output

Invalid Input

```python
import time
t1 = time.time()
print('Input')
a = int(input())
b = int(input())
c = int(input())
print('Output')
list1 = []
if a > 0 and c!=0 and b <1500000 and a < b:
    for x in range(a,b):
        for i in range(2,x):
            if x%i == 0:
                break
        else:
            list1.append(x)
    try:
        print(list1[c])
    except IndexError:
        print('No prime number is present at this index')
else:
    print('Invalid Input')
```
